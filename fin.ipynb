```python
from statsmodels.tsa.stattools import adfuller
```

```python
import yfinance as yf
import pandas as pd
```

```python
msft = yf.Ticker("MSFT")
```

```python
df = pd.DataFrame(msft.history())
```

```python
df
```

```python
chuppe = []
```

```python
print(len(df))
```

```python
for i in range (len(df)):
    chuppe.append(((df["Close"][i]-df["Open"][i])/df["Open"][i]))
    
```

```python
chuppe
```

```python
a = adfuller(chuppe)
a
```

```python
if(a[1]>=0.05):
    print("Since the p-value is not less than .05, we fail to reject the null hypothesis. This means the time series is non-stationary. In other words, it has some time-dependent structure and does not have constant variance over time.")
else:
    print("Since the p-value is less than .05, we accept the null hypothesis. This means the time series is stationary. In other words, it doesn't have some time-dependent structure and does not have constant variance over time.")
```
